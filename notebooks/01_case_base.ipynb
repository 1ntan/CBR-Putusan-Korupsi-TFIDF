Ekstraksi PDF → TXT

```python
import os
import pdfplumber

data = []

for file in uploaded.keys():

    if file.lower().endswith(".pdf"):

        text = ""

        try:

            with pdfplumber.open(file) as pdf:

                for page in pdf.pages:

                    chars = []

                    for c in page.chars:

                        if c["size"] < 20:

                            chars.append(c)

                    page_text = "".join(
                        c["text"]
                        for c in chars
                    )

                    text += page_text + "\n"

            data.append({

                "nama_file": file,
                "isi_putusan": text

            })

            print(f"✓ {file}")

        except Exception as e:

            print(f"✗ {file} -> {e}")
```

Bentuk DataFrame

```python
import pandas as pd

df = pd.DataFrame(data)

print(df.shape)

df.head()
```

cleaning.

```python
import re

def clean_putusan(text):

    text = str(text)

    text = re.sub(
        r"Direktori Putusan Mahkamah Agung Republik Indonesia",
        "",
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r"putusan\.mahkamahagung\.go\.id",
        "",
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r"Halaman\s+\d+\s+dari\s+\d+",
        "",
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r"Disclaimer.*?Telp\s*:\s*021-384\s*3348\s*\(ext\.318\)",
        "",
        text,
        flags=re.IGNORECASE | re.DOTALL
    )

    text = re.sub(
        r"\s+",
        " ",
        text
    )

    return text.strip()
```

Apply Cleaning

```python
df["isi_putusan_clean"] = (
    df["isi_putusan"]
    .apply(clean_putusan)
)

print(
    df.iloc[0]["isi_putusan_clean"][:3000]
)
```

Cek hasil.

```python
for i in range(3):

    print("="*100)

    print(
        df.iloc[i]["nama_file"]
    )

    print(
        df.iloc[i]["isi_putusan_clean"][:1500]
    )
```

```python
df["panjang_clean"] = (
    df["isi_putusan_clean"]
    .str.len()
)

print(
    df["panjang_clean"]
    .describe()
)
```
